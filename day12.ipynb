```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'Year': [2018, 2019, 2020, 2021],
    'Sales': [100, 200, 150, 175],
    'Profit': [30, 50, 45, 60]
}

df = pd.DataFrame(data)

# Line plot (default)
df.plot(x='Year', y='Sales', kind='line', title='Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
# Bar Plot
df.plot(x='Year', y='Sales', kind='bar', title='Sales per Year')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()

# Pie Chart
df.set_index('Year')['Sales'].plot(kind='pie', autopct='%1.1f%%', title='Sales Distribution')
plt.ylabel('')  # Hides the y-label
plt.show()

# Histogram
df['Sales'].plot(kind='hist', bins=5, title='Sales Histogram')
plt.xlabel('Sales')
plt.show()
```

```python
df.plot(x='Year', y='Sales', kind='bar', color='skyblue', title='Sales per Year', grid=True)
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
# Create subplots
df.plot(x='Year', y=['Sales', 'Profit'], kind='line', subplots=True, layout=(2, 1), figsize=(10, 6))
plt.show()
```

```python
df.plot(x='Year', y=['Sales', 'Profit'], kind='line', title='Sales and Profit Over Time')
plt.xlabel('Year')
plt.ylabel('Value')
plt.show()
```
